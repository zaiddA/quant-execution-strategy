```python
def simulate_slippage(row, X, side='buy'):
    """
    row : pandas Series containing:
      ask_px_01…ask_px_10, ask_sz_01…ask_sz_10  (for buys)
      bid_px_01…bid_px_10, bid_sz_01…bid_sz_10  (for sells)
      mid : the mid‐price we computed
    X   : total shares you want to trade
    side: 'buy' or 'sell'
    """
    remaining = X
    cost = 0.0
    # decide which prefixes to use
    price_cols = [f"{side == 'buy' and 'ask' or 'bid'}_px_{i:02d}" for i in range(1,11)]
    size_cols  = [f"{side == 'buy' and 'ask' or 'bid'}_sz_{i:02d}" for i in range(1,11)]

    for pcol, scol in zip(price_cols, size_cols):
        level_sz = row[scol]
        level_px = row[pcol]
        if remaining <= level_sz:
            cost += remaining * level_px
            remaining = 0
            break
        else:
            cost += level_sz * level_px
            remaining -= level_sz

    if remaining > 0:
        # not enough depth → assume infinite slippage or drop these samples
        return None, None

    avg_px    = cost / X
    slippage  = avg_px - row['mid'] if side=='buy' else row['mid'] - avg_px
    return avg_px, slippage
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# ── 1. Prepare your empirical (X, slippage) data ─────────────────────────────

# Suppose you have lists of trade sizes and corresponding slippages:
sizes     = [10, 50, 100, 200, 500]        # X values you tested
slippages = [0.205, 0.205, 0.205, 0.22, 0.598]  # resulting g_t(X) values

# Convert them into NumPy arrays for sklearn:
X_array = np.array(sizes).reshape(-1, 1)
#   - np.array(sizes) makes a 1D array: [10,50,…]
#   - .reshape(-1,1) turns it into a 2D column vector so sklearn sees each size as one sample.

y_array = np.array(slippages)
#   - A 1D array of target values (slippage) corresponding to each size.

# ── 2. Fit the linear model g(X) = beta * X ───────────────────────────────────

lin_model = LinearRegression()
#   - Instantiate an ordinary least squares regressor.

lin_model.fit(X_array, y_array)
#   - Fit the model to your data: finds the slope beta and intercept.
#   - By default, it fits: y = intercept + coef * X.

beta = lin_model.coef_[0]
#   - coef_ is an array of coefficients (one per feature). Here there's only one feature, so index 0.

intercept_lin = lin_model.intercept_
#   - intercept_ is the constant term. For a pure proportional model you might force intercept=0,
#     but here we allow a small offset if it improves fit.

r2_lin = lin_model.score(X_array, y_array)
#   - score() returns R²: the fraction of variance in y explained by the model.

print(f"Linear fit: slippage ≈ {beta:.6f}*X + {intercept_lin:.6f}")
print(f" → R² = {r2_lin:.4f}")

# ── 3. Fit the power-law model g(X) = alpha * X^gamma via log–log regression ───

# We need to take logs, but only where slippage > 0:
mask = y_array > 0
#   - Boolean mask to filter out any zero or negative slippage values (cannot take log of those).

logX = np.log(X_array[mask]).reshape(-1, 1)
#   - Natural log of X: log(X). Reshaped for sklearn input.

logY = np.log(y_array[mask])
#   - Natural log of slippage: log(g).

pow_model = LinearRegression()
pow_model.fit(logX, logY)
#   - Now we fit: logY = a + b * logX  ⇒  g ≈ exp(a) * X^b

gamma = pow_model.coef_[0]
#   - This b is our exponent gamma.

alpha = np.exp(pow_model.intercept_)
#   - exp(a) recovers the multiplicative constant alpha.

r2_pow = pow_model.score(logX, logY)
#   - R² on the log–log regression: how well the power law explains variance in log space.

print(f"Power-law fit: slippage ≈ {alpha:.6f} * X^{gamma:.4f}")
print(f" → R² (log–log) = {r2_pow:.4f}")

# ── 4. Visualize both fits against empirical data ────────────────────────────────

import matplotlib.pyplot as plt

# Scatter the empirical points:
plt.scatter(sizes, slippages, marker='o', label='Empirical', zorder=3)
#   - marker='o' draws circles; zorder=3 keeps points on top.

# Create a smooth range of X for plotting the curves:
xs = np.linspace(min(sizes), max(sizes), 200)

# Plot the linear fit line:
plt.plot(xs, beta*xs + intercept_lin, label='Linear fit', linewidth=2)
#   - linewidth=2 for visibility.

# Plot the power-law curve:
plt.plot(xs, alpha * xs**gamma, label='Power-law fit', linewidth=2)

plt.xscale('log')
plt.yscale('log')
#   - Log–log scales help see both small and large values clearly.

plt.xlabel('Trade size X')
plt.ylabel('Slippage $g_t(X)$')
plt.title('Empirical vs. Fitted Impact Curves')
plt.legend()
plt.grid(True, which='both', ls='--', lw=0.5)
#   - A light dashed grid on both major/minor ticks.

plt.show()
```

```python
import numpy as np

# your data
sizes     = np.array([10, 50, 100, 200, 500])
slippages = np.array([0.205, 0.205, 0.205, 0.220, 0.598])

# Linear fit (force through origin)
beta = np.polyfit(sizes, slippages, deg=1)[0]
pred_lin = beta * sizes
# R² for linear
ss_res_lin = np.sum((slippages - pred_lin)**2)
ss_tot     = np.sum((slippages - slippages.mean())**2)
r2_lin     = 1 - ss_res_lin/ss_tot

# Power‐law fit via log–log
logX = np.log(sizes)
logY = np.log(slippages)
gamma, log_alpha = np.polyfit(logX, logY, deg=1)
alpha = np.exp(log_alpha)
pred_pow = alpha * sizes**gamma
# R² for power‐law (in linear space)
ss_res_pow = np.sum((slippages - pred_pow)**2)
r2_pow     = 1 - ss_res_pow/ss_tot

print(f"Linear:   β = {beta:.6f}, R² = {r2_lin:.4f}")
print(f"PowerLaw: α = {alpha:.6f}, γ = {gamma:.4f}, R² = {r2_pow:.4f}")
```

```python
import numpy as np

def get_level_columns(row, side='buy'):
    """
    Return sorted lists of price- and size-column names for the given side.
    'buy' → ask, 'sell' → bid
    """
    prefix = 'ask' if side == 'buy' else 'bid'
    # Collect candidate columns
    price_candidates = [c for c in row.index if c.startswith(f"{prefix}_px") or c.startswith(f"{prefix}_price")]
    size_candidates  = [c for c in row.index if c.startswith(f"{prefix}_sz") or c.startswith(f"{prefix}_size")]

    # Helper to extract numeric level suffix
    def level_num(col):
        suf = ''.join(filter(str.isdigit, col.split('_')[-1]))
        return int(suf)

    # Sort by level number
    price_cols = sorted(price_candidates, key=level_num)
    size_cols  = sorted(size_candidates,  key=level_num)

    return price_cols, size_cols

def simulate_slippage(row, X, side='buy'):
    """
    Simulate executing a market order of size X on a single snapshot.
    Returns (avg_price, slippage) or (nan, nan) if depth is insufficient.
    """
    price_cols, size_cols = get_level_columns(row, side)
    remaining = X
    cost = 0.0

    for pcol, scol in zip(price_cols, size_cols):
        lvl_size  = row[scol]
        lvl_price = row[pcol]
        if remaining <= lvl_size:
            cost += remaining * lvl_price
            remaining = 0
            break
        else:
            cost += lvl_size * lvl_price
            remaining -= lvl_size

    if remaining > 0:
        # Order not fully filled
        return np.nan, np.nan

    avg_price = cost / X
    # Ensure you’ve computed df['mid'] beforehand
    if side == 'buy':
        slippage = avg_price - row['mid']
    else:
        slippage = row['mid'] - avg_price

    return avg_price, slippage

# Example:
# row0 = snap.iloc[0]
# avg_px, slip = simulate_slippage(row0, 100, side='buy')
# print(f"Executed @ {avg_px:.4f}, slippage = {slip:.4f}")
```

```python
import numpy as np
import pandas as pd

def fit_powerlaw(sizes, slippages):
    """
    Fit g(X) = α X^γ via log–log OLS and return (α, γ).
    """
    mask = slippages > 0
    if mask.sum() < 2:
        return np.nan, np.nan
    logX = np.log(sizes[mask])
    logY = np.log(slippages[mask])
    gamma, log_alpha = np.polyfit(logX, logY, deg=1)
    return np.exp(log_alpha), gamma

# ── Sample and fit across multiple snapshots ───────────────────────────────────
# Take M random snapshots to average out noise
M = 100
snapshots = snap.sample(n=M, random_state=42)

results = []
sizes = np.array([10, 50, 100, 200, 500])

for _, row in snapshots.iterrows():
    # 1) Simulate slippage for each size
    slips = np.array([ simulate_slippage(row, X, side='buy')[1] for X in sizes ], dtype=float)
    valid = ~np.isnan(slips)
    if valid.sum() < 3:
        continue

    # 2) Trim out the top/bottom 5% of slippages in this snapshot
    vals = slips[valid]
    lo, hi = np.quantile(vals, [0.05, 0.95])
    trimmed = np.clip(vals, lo, hi)

    # 3) Fit (α, γ) on the trimmed data
    α, γ = fit_powerlaw(sizes[valid], trimmed)
    if np.isnan(α) or np.isnan(γ):
        continue

    # 4) Record timestamp hour for intraday bucketing
    hour = row['ts'].hour
    results.append({'alpha': α, 'gamma': γ, 'hour': hour})

# Build DataFrame of all snapshot‐level fits
df_res = pd.DataFrame(results)

# ── Aggregate overall median ───────────────────────────────────────────────────
alpha_med = df_res['alpha'].median()
gamma_med = df_res['gamma'].median()
print(f"Overall impact model: α = {alpha_med:.4f}, γ = {gamma_med:.4f}")

# ── Aggregate per‐hour medians ─────────────────────────────────────────────────
hourly = (
    df_res
    .groupby('hour')[['alpha','gamma']]
    .median()
    .reset_index()
    .sort_values('hour')
)
print("\nPer-hour median (α, γ):")
print(hourly.to_string(index=False))
```

```python
import numpy as np

def fit_powerlaw(sizes, slippages):
    mask = slippages > 0
    if mask.sum() < 2:
        return np.nan, np.nan
    logX = np.log(sizes[mask])
    logY = np.log(slippages[mask])
    gamma, log_alpha = np.polyfit(logX, logY, deg=1)
    return np.exp(log_alpha), gamma

# ── Sample and fit ────────────────────────────────────────────────────────────
sizes = np.array([10, 50, 100, 200, 500])
alpha_list = []
gamma_list = []

for _, row in snap.sample(n=100, random_state=42).iterrows():
    # 1) simulate slippages for all 5 sizes
    slips = np.array([simulate_slippage(row, X, 'buy')[1] for X in sizes], dtype=float)
    valid = ~np.isnan(slips)
    if valid.sum() < 3:
        continue

    # 2) trim extremes
    vals = slips[valid]
    lo, hi = np.quantile(vals, [0.05, 0.95])
    trimmed = np.clip(vals, lo, hi)

    # 3) fit on the **corresponding** sizes
    sizes_valid = sizes[valid]
    alpha_val, gamma_val = fit_powerlaw(sizes_valid, trimmed)
    if np.isfinite(alpha_val) and np.isfinite(gamma_val):
        alpha_list.append(alpha_val)
        gamma_list.append(gamma_val)

# ── Compute medians ────────────────────────────────────────────────────────────
if alpha_list:
    alpha_med = np.median(alpha_list)
    gamma_med = np.median(gamma_list)
    print(f"Overall impact model: alpha = {alpha_med:.4f}, gamma = {gamma_med:.4f}")
else:
    print("No valid fits; check simulate_slippage outputs.")
```
